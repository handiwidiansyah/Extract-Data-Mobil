```python
from urllib.request import urlopen
merek = "toyota"
html = urlopen("https://www.olx.co.id/mobil-bekas_c198?filter=make_eq_mobil-bekas-"+merek).read()
print(type(html))
print(html[1:1000])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

```python
items = soup.find_all("div","_2v8Tq")
for i in items:
    print (i.find("div","_2Gr10").get_text())
    print (i.find("span","_1zgtX").get_text())
```

```python
import pandas as pd

merek_mobil = []
harga_mobil = []

items = soup.find_all("div","_2v8Tq")

for i in items:
    nama = i.find("div","_2Gr10").get_text()
    harga = i.find("span","_1zgtX").get_text().replace("Rp ","").replace(".","")
    if harga != "":
        merek_mobil.append(nama)
        harga_mobil.append(harga)
    
dict_items = {"merek":merek_mobil, "harga":harga_mobil}
df = pd.DataFrame(dict_items, columns = ["merek","harga"])
df
```
